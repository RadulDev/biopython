```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
potential = np.genfromtxt([i for i in open('potential.xvg').read().splitlines()
                          if not i.startswith(('#','@'))])
potential
```

```python
plt.plot(*potential.T)
plt.xlabel('time step')
plt.ylabel('potential')
```

```python
temperature = np.genfromtxt([i for i in open('temperature.xvg').read().splitlines()
                          if not i.startswith(('#','@'))])
temperature
```

```python
plt.plot(*temperature.T)
plt.xlabel('time step')
plt.ylabel('temperature')
```

```python
pressure = np.genfromtxt([i for i in open('pressure.xvg').read().splitlines()
                          if not i.startswith(('#','@'))])
pressure
plt.plot(*pressure.T)
plt.xlabel('time step')
plt.ylabel('pressure')
```

```python
denisity = np.genfromtxt([i for i in open('density.xvg').read().splitlines()
                          if not i.startswith(('#','@'))])
denisity
plt.plot(*denisity.T)
plt.xlabel('time step')
plt.ylabel('density')
```

```python
df1 = pd.DataFrame(temperature, columns=['time','temperature'])
df2 = pd.DataFrame(pressure, columns=['time','pressure'])
df3 = pd.DataFrame(denisity, columns=['time','density'])
# df1
# df2
# df3
```

```python
df_main = pd.merge(df1,df2,on='time')
df_main = pd.merge(df_main,df3,on='time')
df_main
```

```python
plt.plot(df_main)
```

```python
plt.plot(df_main['time'],df_main['temperature'])
```

```python
df_main = df_main.set_index('time')
df_main
```

```python
fig = plt.plot(df_main)
# fig.set_dpi(300)
plt.legend(df_main)
```
